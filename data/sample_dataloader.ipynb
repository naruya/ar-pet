```python
for i in range(len(glob.glob("dataset/aux/*.pkl"))):
    aux_files = sorted(glob.glob("dataset/aux/*.pkl"))[:i+1]
    auxs = [pickle.load(open(file, 'rb')) for file in aux_files]
    auxs = np.array(list(chain.from_iterable(auxs)))

    img_files = sorted(glob.glob("dataset/img/*.gif"))[:i+1]
    imgs = [vread(file) for file in img_files]
    imgs = np.array(list(chain.from_iterable(imgs)))
    print(i, auxs.shape[0] == imgs.shape[0], auxs.shape[0], imgs.shape[0])
```
